```python
# Importando as bibliotecas necessarias 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('autos.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["vehicleType"].value_counts().plot.pie(ylabel='', title='Tipos de VeÃ­culos')
```

```python
df["gearbox"].value_counts().plot.pie(ylabel='', title='CÃ¢mbio utilizado')
```

```python
df["fuelType"].value_counts().plot.bar(ylabel='', title='Combustivel utilizado')
```

```python
df.groupby(['yearOfRegistration']).mean().plot(y='price')
plt.legend()
plt.title('PreÃ§o mÃ©dio ao longo dos anos')
```

```python
df.groupby(['yearOfRegistration']).mean().plot(y='powerPS')
plt.legend()
plt.title('MÃ©dia da PotÃªncia de motor ao longo do tempo')
```

```python
df.groupby(['brand']).mean().plot(kind='bar', y='powerPS', figsize=(10,6))
plt.legend()
plt.title('MÃ©dia da PoÃªncia por Marca')
```
