```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataframe = pd.read_csv('./ead.csv')
```

```python
dataframe.head(n=6)
```

```python
dataframe.dtypes
```

```python
dataframe["Ano"] = dataframe["Ano"].astype(float).astype(str)
```

```python
dataframe["Posicao_P_Google"] = dataframe["Posicao_P_Google"].astype(float).astype(str)
```

```python
dataframe.dtypes
```

```python
dataframe.head(n=6)
```

```python
dataframe["grau_academico"].value_counts().plot.pie(ylabel='')
```

```python
dataframe["eixo"].value_counts().plot.pie(ylabel='')
```

```python
dataframe["Curso"].value_counts().sort_values().plot.bar(title="")
```

```python
df = pd.DataFrame(dataframe,columns=['Curso','Matricula'])
ax = df.plot.bar(x = 'Curso', y= 'Matricula')
```

```python
df = pd.DataFrame(dataframe,columns=['Curso','Matricula'])

df.groupby(['Curso']).sum().plot(kind='pie', y='Matricula', ylabel='')
plt.legend().remove()
```

```python
df = pd.DataFrame(dataframe,columns=['eixo','Matricula'])

df.groupby(['eixo']).sum().plot(kind='pie', y='Matricula', ylabel='')
plt.legend().remove()
```

```python
df = pd.DataFrame(dataframe,columns=['Ano','Matricula'])
sns.lineplot(x="Ano", y="Matricula", data=df)
plt.xticks(rotation=15)
plt.title('Matricula dos Top 10 durante os anos ')
plt.show()
```

```python
df = pd.DataFrame(dataframe,columns=['Ano','eixo','grau_academico'])
#plot multiple time series
sns.lineplot(x='Ano', y='eixo', hue='grau_academico', data=df)
```

```python
df = pd.DataFrame(dataframe,columns=['Ano','Curso','Posicao_P_Google'])
#plot multiple time series
sns.lineplot(x='Ano', y='Curso', hue='Posicao_P_Google', data=df)
plt.legend([],[], frameon=False)
```

```python
dataframe = pd.DataFrame(dataframe,columns=['Matricula', 'Posicao_P_Google'])

# Load data into DataFrame
df = pd.DataFrame(dataframe);

_ = sns.lmplot(x = 'Matricula', y= 'Posicao_P_Google', data=df, ci=None)
```
