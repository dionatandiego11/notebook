```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dados = pd.read_csv('./dados.csv') 
```

```python
dados.head(n=6)
```

```python
Total = dados['valorPagoAtualizado'].sum()
print ("Total Pago:",Total)
```

```python
df = pd.DataFrame(dados,columns=['descricaoFuncao','valorPagoAtualizado'])

df.groupby(['descricaoFuncao']).sum().plot(kind='bar', y='valorPagoAtualizado', ylabel='', title='Gasto por segmento')
plt.legend().remove()
```

```python
df = pd.DataFrame(dados,columns=['descricaoUnidade','valorPagoAtualizado'])

df.groupby(['descricaoUnidade']).sum().plot(kind='bar', y='valorPagoAtualizado', ylabel='', title='Origem das receitas por secretaria')
plt.legend().remove()
```
